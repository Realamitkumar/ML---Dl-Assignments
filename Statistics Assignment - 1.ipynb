# **Problem Statement 8:**

- I work as a data analyst in Aeon Learning Pvt. Ltd. After analyzing data,
- I make reports, where I have the efficiency of entering 77 words
- per minute with 6 errors per hour.

- What is the probability that I will commit 2 errors in a 455-word financial report?

- What happens when the no. of words increases/decreases (in case of 1000 words, 255 words)?
- How is the  affected? How does it influence the PMF?

Give a pictorial representation of the same to validate your answer.

```python
import numpy as np
import matplotlib.pyplot as plt 

n_errors_per_min=6/60 
words=[100,255,455,1000,1500]
prob=[]
for i in words:
    lamda=(i/77)*n_errors_per_min
    dist=np.random.poisson(lamda,1000)
    prob.append(np.sum(dist==2)/1000)
plt.plot(words,prob)
plt.xlabel("Number of Words")
plt.ylabel("Probability of two errors");
```

# OR

```python
# What is the probability that I will commit 2 errors in a 455-word financial report?
# in an interval of 455 words 
import math
u = 455 * 0.0012
prob1 = round(math.exp(-u) * math.pow(u,2) / math.factorial(2),2)
prob1
```

```python
# What happens when the no. of words increases/decreases (in case of 1000 words,255 words)?
u = 1000 * 0.0012
prob_1000 = round(math.exp(-u) * math.pow(u,2) / math.factorial(2),2)

u = 255 * 0.0012
prob_255 = round(math.exp(-u) * math.pow(u,2) / math.factorial(2),2)

print(prob_1000, prob_255)
```
