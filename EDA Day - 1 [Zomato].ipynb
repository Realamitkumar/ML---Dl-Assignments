```python
#Missing Value  - 1 
df.isnull().sum()
```

```python
#Missing Value  - 2 [list_comprehension]
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
#Missing Value  - 3 [function]
lis  =  []
for i in df.columns:
    if df[i].isnull().sum()>0:
        lis.append(i)
  
lis
```

```python
#Missing Value  - 1 [heatmap]
#hear nothing will be visible, just becoz missing
#values are very very small
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis');
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns #one common column that is 'Country Code'
```

```python
final_df = pd.merge(df, df_country,on = 'Country Code',how='left')
final_df.head(2)
```

```python
## To check data type
final_df.dtypes
```

```python
final_df.columns
```

``Pie-Chart``

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
### Pie - Chart- Top 3 countries that uses Zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%");
```

- Observation : 
- ``Zomato`` Maximum records or transaction are from India
- After that US and UK 

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (14,8)
sns.barplot(x = 'Aggregate rating',y = 'Rating Count',data = ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green']);
```

# Observation :
1. Not Rated Count is very high
2. Maximum Number Of Rating are between 2.5 to 3.4

```python
# Count Plot
import seaborn as sns
sns.countplot(x = 'Rating color', data = ratings,palette=['blue','red','orange','yellow','green','green']);
```

```python
#countries name that has given 0 rating
#1
c1=final_df[final_df['Aggregate rating']==0.0].groupby('Country').size().reset_index()

#2
c=final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```
